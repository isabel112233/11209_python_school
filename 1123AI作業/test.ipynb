```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from datasets import Dataset
```

```python
matplotlib.rc('font', family='Microsoft JhengHei')
df = pd.read_csv('./地區房價資料檔.csv')
featuresCorr = df.corr()
featuresCorr
```

```python
matplotlib.rc('font', family='Microsoft JhengHei')
sns.heatmap(featuresCorr)
sns.displot(featuresCorr,kde=True)
```

```python
targetCorr = featuresCorr['房價']
targetCorr = targetCorr.drop('房價')
selectedFeatures = targetCorr[abs(targetCorr) > 0.4]
print("選擇特徵數： {} \n選擇特徵:\n{}".
      format(len(selectedFeatures), selectedFeatures))
```

```python
matplotlib.rc('font', family='Microsoft JhengHei')
df = pd.read_csv("./地區房價資料檔.csv")
df = pd.DataFrame(df)
df
```

選取1低收入比、2房間數、3師生比為特徵
